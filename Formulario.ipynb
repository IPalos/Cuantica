## Operador Hermitiano, y sus propiedades

### Definición

Un operador se llama hermitiano si cumple la condición
\begin{align}
\left( \phi, \hat{A} \psi \right) = \left( \hat{A}\phi,\psi \right)
\end{align}
Es decir, que : $\hat{B} = \hat{\dagger{B}}$ 

### Teorema

Si $\hat{A}$ es hermitiano, entonces sus eigenvalores son reales 

### Teorema

Si $\hat{A}$ es hermitiano y tenemos una colección de eigenvectores $\chi_i$, si $a_n \ne a_m$ entonces $\chi_n \perp \chi_m$

### Teorema
Dos operadores $\hat{A}$ y $\hat{B}$ conmutan si y sólo si $\exists$ un conjunto de eigenvalores comunes

## El Conmutador

\begin{align}
\boxed{[\hat{A} ,\hat{B} ] \equiv \hat{A}\hat{B} - \hat{B}\hat{A} }
\end{align}

## Pozo Infinito de longitud L

### Hamiltoniano

\begin{align}
\boxed{\hat{\mathcal{H}} = \frac{-\hbar^2}{2m}\frac{d}{dx^2}}
\end{align}

### Función de onda

\begin{align}
\boxed{\psi(x) = \sqrt{\frac{2}{L}}sen\left( \frac{\pi n }{L}x \right)}
\end{align}

### Energías

\begin{align}
\boxed{E_n = \frac{\pi^2 \hbar^2}{2m L^2}n^2}
\end{align}

## Varianza

\begin{align}
\boxed{\sigma^2_\psi [x] =\int_{-\infty}^{\infty} (x-\langle x\rangle )^2 |\psi | ^2 dx}
\end{align}

Definimos a $\delta\hat{x}$ como 

\begin{align}
\delta\hat{x} = \hat{x}- \langle x \rangle \hat{I}
\end{align}

### Propiedades 
- $[\delta\hat{A},\delta\hat{B}] = [\hat{A},\hat{B}]$
- Si $\hat{A}$ es hermitiano, $\delta{\hat{A}}$ es autoadjunto




## Desigualdades de Heisenberg

\begin{gather}
\boxed{\frac{1}{2}\left|\langle[\hat{A},\hat{B}]\rangle \right| \le ||\delta\hat{A}\psi|| \; ||\delta\hat{B}\psi|| \\ \\ \\ \Delta A \Delta B \ge \frac{1}{2}|\langle [\hat{A},\hat{B}]\rangle|
}
\end{gather}
